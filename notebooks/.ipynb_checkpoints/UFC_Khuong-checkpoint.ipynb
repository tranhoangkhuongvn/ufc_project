### Data Cleaning:
1. For fighters dataset
- Convert Height to cms, Weight to kg
- Convert Reach and Stance to numerical values
- Extract DOB to get year --> get age of fighter

2. For matches dataset:
- Convert date to datetime type

3. Merge fighters data to match data

```python
# Since height and weights are object types, we need to convert them to numbers, 
# TODO: convert height to cm and weight to kg
fighters_df.dtypes
```

```python
def convert_height(string):
    """
    input is a string of 5' 10" or NaN
    output is height in cms
    """
    if pd.isna(string):
        return None
    new_string = string.split(" ")
    feet = int(new_string[0].replace("\'", ''))
    inches = int(new_string[1].replace('"', '')  )
    return feet * 30.48 + inches * 2.54
    
#convert_height('''5' 10"''') # output should be 177.8
```

```python
fighters_df.Height[0]
```

```python
fighters_df.Height = fighters_df.Height.apply(convert_height)
fighters_df.dtypes
```

```python
# Percentage of missing data per column
fighters_df.isnull().sum() 
```

```python
#### Convert weight to kg
def convert_weight(string):
    """
    input: string contains weight in lbs
    output: weight in kg
    """
    if pd.isna(string):
        return None
    weight = string.split(" ")[0]
    return int(weight) * 0.453592
```

```python
fighters_df.Weight = fighters_df.Weight.apply(convert_weight)
```

```python
fighters_df.dtypes
```

```python
def convert_reach(string):
    """
    input: NaN or 76'' (inches)
    output: number in cms
    """
    if pd.isna(string):
        return string
    reach = string.replace('"', "")
    return int(reach) * 2.54
    
```

```python
fighters_df.Reach = fighters_df.Reach.apply(convert_reach)
```

```python
fighters_df.dtypes
```

```python
# Extract year from DOB
def extract_year(string):
    if pd.isna(string):
        return string
    year = string.split(" ")[2]
    return int(year)
```

```python
fighters_df['year'] = fighters_df.DOB.apply(extract_year)
```

#### Merge fighters and matches df 

```python
r_merge_df = matches_df.merge(fighters_df, left_on='R_fighter', right_on='fighter_name', how='left')
r_merge_df.rename(columns={'Height':'R_Height',
                          'Weight':'R_Weight',
                          'Reach':'R_Reach',
                          'Stance':'R_Stance',
                          'DOB':'R_DOB',
                          'year': 'R_year'}, 
                 inplace=True)
```

```python
r_merge_df.drop('fighter_name', axis=1, inplace=True)
r_merge_df.head()
```

```python
b_merge_df = r_merge_df.merge(fighters_df, left_on='B_fighter', right_on='fighter_name', how='left')
b_merge_df.rename(columns={'Height':'B_Height',
                          'Weight':'B_Weight',
                          'Reach':'B_Reach',
                          'Stance':'B_Stance',
                          'DOB':'B_DOB',
                          'year': 'B_year'}, 
                 inplace=True)
```

```python
b_merge_df.drop('fighter_name', axis=1, inplace=True)
b_merge_df.head()
```

```python
b_merge_df.isnull().sum().sort_values()
```

```python
b_merge_df.to_csv("merged_df.csv", index=False)
```

#### Get the age of the fighters by the time they fight

```python
b_merge_df['R_age'] = b_merge_df['date'].dt.year - b_merge_df['R_year']
b_merge_df['B_age'] = b_merge_df['date'].dt.year - b_merge_df['B_year']
b_merge_df.head()
```

```python
b_merge_df.to_csv("updated_merged_df.csv", index=False)
```
